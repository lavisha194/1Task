```python
# Filter data for total population
total_population_data = df[df['Continent'] == 'Africa']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022 Population", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Capital']])
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022 Population", y="CCA3", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2015 Population", y="CCA3", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2015)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
# Sort data based on the total population for 2022
total_population_sorted1 = total_population_data.sort_values(by="2022 Population", ascending=True)

# Get the top ten countries with the highest total population for 2022
total_bottom_ten_countries = total_population_sorted1.head(10)
print("Bottom ten countries of total population\n")
print(total_bottom_ten_countries[['CCA3']])
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022 Population", y="CCA3", data=total_bottom_ten_countries, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2015 Population", y="CCA3", data=total_bottom_ten_countries, palette="coolwarm")
plt.title("Bottom Ten Countries of Total Population (2015)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
#filter data for male population
male_population_data = df[df['Continent']=='Africa']

#sort data based on the male population for 2022
male_population_sorted =male_population_data.sort_values(by="2022 Population",ascending=False)

#Get the top 10 countries with the highest male population for 2022
male_top_ten_countries = male_population_sorted .head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['CCA3']])
```

```python
#filter data for male population
female_population_data = df[df['Continent']=='Africa']

#sort data based on the male population for 2022
female_population_sorted =female_population_data.sort_values(by="2022 Population",ascending=False)

#Get the top 10 countries with the highest male population for 2022
female_top_ten_countries = female_population_sorted .head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['CCA3']])
```

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022 Population", y="CCA3", data=male_top_ten_countries, palette="viridis")
plt.title("Top ten conuntries of male population(2022)",fontsize=10)
plt.xlabel("Male Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022 Population", y="CCA3", data=female_top_ten_countries, palette="viridis")
plt.title("Top ten conuntries of female population(2022)",fontsize=10)
plt.xlabel("Female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

```python
#merge male and female population data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="CCA3",suffixes=("_male","_female"))
```

```python
#merged data
#calculate the total population for each country (male + female)
merge_data["Total population"] = merge_data["2022 Population_male"] + merge_data ["2022 Population_female"]
```

```python
merge_data.head()
```

```python
#sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

```python
top_10_countries = sorted_data.head(10)
```

```python
#create the staked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="CCA3",y="2022 Population_female",data=top_10_countries,color="red",label="female Population")
sns.barplot(x="CCA3",y="2022 Population_male",data=top_10_countries,bottom=top_10_countries["2022 Population_female"],color="green",label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
#select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="CCA3", y="2022 Population_female", data=bottom_10_countries, color="red", label="Female Population")
sns.barplot(x="CCA3", y="2022 Population_male", data=bottom_10_countries, bottom=bottom_10_countries["2022 Population_female"], color="green", label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
